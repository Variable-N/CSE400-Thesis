```python
from sklearn.cluster import KMeans
import numpy as np
import pickle
```

```python
with open('xnxx.pickle', 'rb') as handle:
    train_sents = pickle.load(handle)
```

```python
x = []
for sent in train_sents:
  for word in sent:
    x.append(word)
```

```python
x[0].shape
```

```python
num_clusters = 25
model = KMeans(n_clusters=num_clusters)
model.fit(x)
cluster_assignment = model.labels_
```

```python
cluster_assignment 
```

```python
model.predict([x[0]])
```
