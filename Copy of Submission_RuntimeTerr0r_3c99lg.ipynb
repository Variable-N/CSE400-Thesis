```python
def gender_encoder(x):
    if x == 'M': return 1
    elif x == 'F': return 0

df['gender']=df.gender.apply(lambda x: gender_encoder(x))
```

```python
def YN_encoder(x):
    if x == 'Y': return 1
    elif x == 'N': return 0

df['s11'] = df.s11.apply(lambda x: YN_encoder(x))
df['s12'] = df.s12.apply(lambda x: YN_encoder(x))
```

```python
df.head()
```

```python
def ABCD_encoder(x):
    if x == 'A' or x == 'a': return 0.25
    elif x == 'B' or x == 'b': return 0.5
    elif x == 'C' or x == 'c': return 0.75
    elif x == 'D'or x == 'd': return 1

df['s16'] = df.s16.apply(lambda x: ABCD_encoder(x))
df['s17'] = df.s17.apply(lambda x: ABCD_encoder(x))
df['s18'] = df.s18.apply(lambda x: ABCD_encoder(x))
df['s71'] = df.s71.apply(lambda x: ABCD_encoder(x))
```

```python
def s52_encoder(x):
    if x == 'o': return 0.2
    elif x == 'l': return 0.8
    else: return x

df['s52'] = df.s52.apply(lambda x: s52_encoder(x))
```

```python
def s54_encoder(x):
    if x == '2b' : return -1
    elif x == 'b2': return -0.75
    elif x == 'a2': return -0.5
    elif x == '22': return -0.25
    elif x == 'ab': return 1
    elif x == '2a': return 0.75
    elif x == 'ba': return 0.5
    elif x == 'bb': return 0.25
    else: return 0.0

df['s54'] = df.s54.apply(lambda x: s54_encoder(x)).astype('float')
```

```python
def s55_encoder(x):
    if x == '2k' : return -1
    elif x == '2K': return -0.75
    elif x == 'k2': return -0.5
    elif x == 'K2': return -0.25
    elif x == '22': return 1
    elif x == 'Kk': return 0.75
    elif x == 'KK': return 0.5
    elif x == 'kk': return 0.25
    elif x == 'kK': return 1.25
    else: return 0.0

df['s55'] = df.s55.apply(lambda x: s55_encoder(x)).astype('float')
```

```python
def AB_encoder(x):
    if x == 'A' : return 1
    elif x == 'B': return 0

df['s58'] = df.s58.apply(lambda x: AB_encoder(x))
```

```python
def s69_encoder(x):
    if x == 'x' : return 1
    elif x == '~1': return 0.66
    elif x == 'C`': return 0.33
    else : return 0

df['s69'] = df.s69.apply(lambda x: s69_encoder(x)).astype('float')
```

```python
def s70_encoder(x):
    if x[-1] == 'A' : return 1
    elif x[-1] == 'B': return 0.66
    elif x[-1] == 'C': return 0.33
    else : return 0

df['s70'] = df.s70.apply(lambda x: s70_encoder(x)).astype('float')
```

```python
for col in df:
  print("===================   "+col+"   ====================")
  print(df[col].value_counts())
```

```python
df1 = df.copy()
df1.drop(['id'], axis=1, inplace=True)
x = df1.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
```

```python
df.head()
```

```python
from sklearn.utils import resample
#create two different dataframe of majority and minority class 
df_majority = df[(df[30]==0.0)] 
df_minority = df[(df[30]==1.0)] 
# upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,    # sample with replacement
                                 n_samples= 23562, # to match majority class
                                 random_state=42)  # reproducible results
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_minority_upsampled, df_majority])
```

```python
df_upsampled[30].value_counts()
```

```python
X = df_upsampled.loc[:,0:29]
X
```

```python
Y = df_upsampled[30]
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, shuffle = True)
```

```python
model_LR = LogisticRegression(max_iter=30000).fit(X_train,Y_train)
model_SVM = SVC(kernel='rbf', random_state = 1).fit(X_train, Y_train)
model_NB = ComplementNB().fit(X_train,Y_train)
```

```python
LR_score = classification_report(Y_test, model_LR.predict(X_test))
SVC_score = classification_report(Y_test, model_SVM.predict(X_test))
NB_score = classification_report(Y_test, model_NB.predict(X_test))
print("LOGISTIC REGRESSION")
print(LR_score)
print("SUPPORT VECTOR CLASSIFICATION")
print(SVC_score)
print("NAIVE BAYES CLASSIFICATION")
print(NB_score)
```

```python
import tensorflow as tf
from keras import layers
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from sklearn import preprocessing, decomposition, model_selection, metrics, pipeline
from keras.layers import Flatten, Bidirectional, SpatialDropout1D
from keras.callbacks import EarlyStopping
```

```python
%tensorflow_version 2.x
import tensorflow as tf
print("Tensorflow version " + tf.__version__)

try:
  tpu = tf.distribute.cluster_resolver.TPUClusterResolver()  # TPU detection
  print('Running on TPU ', tpu.cluster_spec().as_dict()['worker'])
except ValueError:
  raise BaseException('ERROR: Not connected to a TPU runtime; please see the previous cell in this notebook for instructions!')

tf.config.experimental_connect_to_cluster(tpu)
tf.tpu.experimental.initialize_tpu_system(tpu)
tpu_strategy = tf.distribute.experimental.TPUStrategy(tpu)
```

```python
X_train.shape
```

```python
with tpu_strategy.scope():
  model = Sequential()
  model.add(Dense(512, input_dim = 30, activation = 'relu'))
  model.add(Dense(256, activation='relu'))
  model.add(Dropout(0.2))
  model.add(Dense(128, activation='relu'))
  model.add(Dense(64, activation='relu'))
  model.add(Dropout(0.1))
  model.add(Dense(32, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(X_train, Y_train, epochs=25, batch_size=256, validation_split=0.1,callbacks=[EarlyStopping(monitor='val_loss', patience=2, min_delta=0.0001)])
```

```python
accr = model.evaluate(X_test,Y_test)
print('Test set\n  Loss: {:0.3f}\n  Accuracy: {:0.3f}'.format(accr[0],accr[1]))
```
